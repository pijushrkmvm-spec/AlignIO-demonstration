# Introduction to Bio.AlignIO
`AlignIO` is Biopython's standard tool for **reading, writing, and manipulating** Multiple Sequence Alignments (MSAs). 

### **Why do we need it?**
Bioinformatics has dozens of alignment formats (FASTA, Clustal, PHYLIP, etc.). `AlignIO` acts as a **universal translator**, letting us handle all these formats using the exact same Python commands.

### **NOTE**
`AlignIO` does not perform alignment.

## Reading alignment using AlignIO

```python
from Bio import AlignIO

alignment = AlignIO.read("demo_protein.aln", "clustal")
print (alignment)
```

## Explore alignment

```python
# Counting number of sequences in alignment
print ("Number of sequence: ",len(alignment))

# Counting number of columns () in alignment
print ("Alignment length: ",alignment.get_alignment_length())

# Printing header and sequence at once
for record in alignment:
    print(record.id)
    print(record.seq)
```

## Column wise analysis

```python
for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    # alignment[:,i] means
    # We are trying to get ith column from all sequences
    # (:) means All Sequences
    
    print (f"Column {i}: {column}")
    # Here we are extracting columns to analyze conservation across sequences
```

## Find conserved positions

```python
for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    # Here also we are trying to get ith column from all sequences
    
    if len(set(column)) == 1: # set(column) --> Unique characters
        
        print(f"Conserved at position {i}: {column}")
```

## Logic behind `len(set(column))`

### `set(column)` gives unique characters

### Case I

```python
x = 'AAA'
set(x)
```

```python
len(set(x))
```

### Case II

```python
x = 'AGA'
set(x)
```

```python
len(set(x))
```

## So `Length = 1` --> Conserved and `Length > 1` --> Not Conserved 

## Slice alignment

```python
sub_align = alignment[:, 10:30]
# Extracting columns 10 to 30 from all (:) sequences in alignment
print(sub_align)
```

## Convert format

```python
AlignIO.write(alignment, "aligned.fasta", "fasta")
with open ('aligned.fasta', 'r') as file:
    f = file.read()
    print(f)
```

# Problem in finding conservation using `AlignIO`

### The AlignIO method of finding conserved site follows very strict elimination method. Even if only one sequence fails to align at a particular position it shows that the alignment is not conserved at that site. But biologically this is not correct.

### To solve this we can take help of another module namely `Counter` from `collections`

### Instead of `Counter`, `dictionary` method can also be used.

```python
from collections import Counter

for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    
    counts = Counter(column)
    # Counter creates the frequency dictionary of each base in each columns of alignment
    # print(counts)
    
    base, count = counts.most_common(1)[0]
    # print(counts.most_common(1))
    # Working principle
    # counts.most_common(1) --> [('A', 3)] ; It returns a list
    # [0] --> ('A', 3)
    # print(counts.most_common(1)[0])
    score = count / len(column)
    
    print(f"Position {i}: {dict(counts)}, Dominant = {base}, Score = {score:.2f}")
```
